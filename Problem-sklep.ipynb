**Problem schedulingu**

Opis problemu:

Celem zadania było tak wypełnić kalendarz pracownikami, aby zminimalizować wypłaty.

Zmienne decyzyjne:

Poszczególni pracownicy

Funkcja celu:

Minimalizacja kosztów wypłat w tygodniu

min(czas pracy * suma(przepracowane dni* płaca)

Ograniczenia: 

Podane w kodzie programu.

```
from pulp import *
import pandas as pd
```

```
from pulp import *

prob = LpProblem("Minimalizacja wypłat dla pracowników",LpMinimize)

A_pon = LpVariable("Ania_pon", 0, 1, LpInteger)
A_wt = LpVariable("Ania_wt", 0, 1, LpInteger)
A_sr = LpVariable("Ania_sr", 0, 1, LpInteger)
A_czw = LpVariable("Ania_czw", 0, 1, LpInteger)
A_pt = LpVariable("Ania_pt", 0, 1, LpInteger)
S_pon = LpVariable("Stefan_pon", 0, 1, LpInteger)
S_wt = LpVariable("Stefan_wt", 0, 1, LpInteger)
S_sr = LpVariable("Stefan_sr", 0, 1, LpInteger)
S_czw = LpVariable("Stefan_czw", 0, 1, LpInteger)
S_pt = LpVariable("Stefan_pt", 0, 1, LpInteger)
H_pon = LpVariable("Hektor_pon", 0, 1, LpInteger)
H_wt = LpVariable("Hektor_wt", 0, 1, LpInteger)
H_sr = LpVariable("Hektor_sr", 0, 1, LpInteger)
H_czw = LpVariable("Hektor_czw", 0, 1, LpInteger)
H_pt = LpVariable("Hektor_pt", 0, 1, LpInteger)
O_pon = LpVariable("Olaf_pon", 0, 1, LpInteger)
O_wt = LpVariable("Olaf_wt", 0, 1, LpInteger)
O_sr = LpVariable("Olaf_sr", 0, 1, LpInteger)
O_czw = LpVariable("Olaf_czw", 0, 1, LpInteger)
O_pt = LpVariable("Olaf_pt", 0, 1, LpInteger)
L_pon = LpVariable("Lidia_pon", 0, 1, LpInteger)
L_wt = LpVariable("Lidia_wt", 0, 1, LpInteger)
L_sr = LpVariable("Lidia_sr", 0, 1, LpInteger)
L_czw = LpVariable("Lidia_czw", 0, 1, LpInteger)
L_pt = LpVariable("Lidia_pt", 0, 1, LpInteger)
P_pon = LpVariable("Piotr_pon", 0, 1, LpInteger)
P_wt = LpVariable("Piotr_wt", 0, 1, LpInteger)
P_sr = LpVariable("Piotr_sr", 0, 1, LpInteger)
P_czw = LpVariable("Piotr_czw", 0, 1, LpInteger)
P_pt = LpVariable("Piotr_pt", 0, 1, LpInteger)


# zarobki za godzinę pracy
A_pln = 100
S_pln = 50
H_pln = 60
O_pln = 40
L_pln = 11
P_pln = 70

# Funkcja celu
prob += (A_pon+A_wt+A_sr+A_czw+A_pt) * A_pln*4 + (S_pon+S_wt+S_sr+S_czw+S_pt) * S_pln*4 + (H_pon+H_wt+H_sr+H_czw+H_pt) * H_pln*4 + (O_pon+O_wt+O_sr+O_czw+O_pt) * O_pln*4 + (L_pon+L_wt+L_sr+L_czw+L_pt) * L_pln*4 + (P_pon+P_wt+P_sr+P_czw+P_pt) * P_pln*4, "minimalizacja"


#maksymalna liczba osób danego dnia
prob += A_pon+S_pon+H_pon+O_pon+L_pon+P_pon == 2, "pon"
prob += A_wt+S_wt+H_wt+O_wt+L_wt+P_wt== 1, "wt"
prob += A_sr+S_sr+H_sr+O_sr+L_sr+P_sr == 1, "sr"
prob += A_czw+S_czw+H_czw+O_czw+L_czw+P_czw == 1, "czw"
prob += A_pt+S_pt+H_pt+O_pt+L_pt+P_pt == 3, "pt"

# Dni w które danego pracownika nie ma w pracy
# Ania i Olaf moze być w pracy codziennie
prob += S_wt+S_sr+S_czw+S_pt == 0 , "Stefanek"
prob += H_czw+H_pt == 0, "Hektorek"
prob += L_pon+L_wt+L_sr == 0, "Lidka"
prob += P_czw+P_pt == 0, "Piotrek"

# maksymalnie 3 dni w tygodniu dla danego pracownika
prob += A_pon+A_wt+A_sr+A_czw+A_pt <= 3, "Ania" 
prob += S_pon+S_wt+S_sr+S_czw+S_pt <= 3, "Stefan" 
prob += H_pon+H_wt+H_sr+H_czw+H_pt <= 3, "Hektor" 
prob += O_pon+O_wt+O_sr+O_czw+O_pt <= 3, "Olaf"
prob += L_pon+L_wt+L_sr+L_czw+L_pt <= 3, "Lidia"
prob += P_pon+P_wt+P_sr+P_czw+P_pt <= 3, "Piotr"

#minimalnie 1 dzień w tygodniu(trzeba dać ludziom zarobić)
prob += A_pon+A_wt+A_sr+A_czw+A_pt >= 1, "Aniaa" 
prob += S_pon+S_wt+S_sr+S_czw+S_pt >= 1, "Stefann" 
prob += H_pon+H_wt+H_sr+H_czw+H_pt >= 1, "Hektorr" 
prob += O_pon+O_wt+O_sr+O_czw+O_pt >= 1, "Olaff"
prob += L_pon+L_wt+L_sr+L_czw+L_pt >= 1, "Lidiaa"
prob += P_pon+P_wt+P_sr+P_czw+P_pt >= 1, "Piotrr"

prob.writeLP("tydzien.lp")

prob.solve()
print("Status:", LpStatus[prob.status])

print("Koszta:", value(prob.objective), "zl\n") 

for v in prob.variables():
    print(v.name, "=", v.varValue)
```
